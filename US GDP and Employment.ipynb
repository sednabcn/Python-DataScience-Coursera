```python
def drop_consecutive_duplicates(x):
    import pandas as pd
    df=x
    rows=df.shape[0]
    ii=1
    while((df.index[ii]!=df.index[ii-1]) & (ii <= rows)):
                   ii+=1
    
    jj=ii
    if jj > rows:
        return print("There are not consecutive duplicates indexes")
    else:
        return pd.concat([df[:jj],df[jj+1:]],axis=0)
    
def make_dashboard(x, gdp_change_current, unemployed_percent,title):
        import matplotlib.pyplot as plt
        from matplotlib import pylab
        from matplotlib.ticker import MaxNLocator
        import mpld3
                      
        fig=plt.figure(figsize=(10,5))
        ax=fig.gca()
        xmin=int(x.min())
        xmax=int(x.max())
        ax.set_xlim(xmin,xmax)
        ax.xaxis.set_major_locator(MaxNLocator(integer=True))
        ax.plot(x, gdp_change_current, color="firebrick",linewidth=4,label="% GDP_change_current")
        ax.plot(x,unemployed_percent,linewidth=4, label="% unemployed")
        ax.set_title(title)        
        pylab.xlabel("year")
        pylab.ylabel("%")
        ax.legend(loc='best') 
        return mpld3.display(fig)   
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 100)
pd.set_option('display.width', 100)

gdp=pd.read_csv('https://datahub.io/core/gdp-us/r/year.csv',index_col=0)
df=gdp
gdp_current=df['change-current']
employment=pd.read_csv('https://datahub.io/core/employment-us/r/aat1.csv',index_col=0)
de=employment[employment.columns[:-1]]
unemd=de.unemployed_percent
unem=drop_consecutive_duplicates(unemd)
year_gdp=gdp_current.index
year_unem=unem.index

year=[name for name in year_gdp if name in year_unem]
gdp_current_year=np.array(gdp_current[year])
pp=unem[year]
unem_year=np.array(unem[year])
x=np.array(year)
title='US GDP and Employment'
make_dashboard(x, gdp_current_year, unem_year,title)
```
